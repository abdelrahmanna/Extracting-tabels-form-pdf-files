```python
import camelot
table1 = camelot.read_pdf('1957.pdf', pages='4', flavor='stream', row_tol=20
                          , edge_tol=25,table_areas=['30,750,550,300'],
                          columns=['142.5,182,215,250,290,325,358,400,435,470,500'],
                        strip_text='\n',
                         split_text= True)
```

```python
df = table1[0].df
```

```python
df
```

```python
import matplotlib.pyplot as plt
camelot.plot(table1[0], kind='contour')
plt.figure(figsize=(50,50))
plt.show()
```

```python
camelot.plot(table1[0], kind='grid')
plt.figure(figsize=(50,50))
plt.show()
```

```python
df.to_csv('table1.csv')
```
